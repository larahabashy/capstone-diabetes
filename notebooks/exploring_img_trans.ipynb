```python
from torchvision import datasets, transforms, utils
```

```python
from PIL import Image
```

```python
img = Image.open('USLH071_02.jpg')
```

```python
img
```

```python
preprocess = transforms.Compose([
   transforms.Resize((256)),
    #transforms.ColorJitter(contrast=0.1),
   transforms.CenterCrop((175, 224)),
   transforms.ToTensor()
])

x = preprocess(img)
y = transforms.functional.adjust_contrast(x, contrast_factor = 2) # if we wanted we would have to define a function transformer
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

reverse_preprocess = transforms.Compose([
    transforms.ToPILImage(),
    np.array,
])

plt.imshow(reverse_preprocess(x));
```

```python
reverse_preprocess = transforms.Compose([
    transforms.ToPILImage(),
    np.array,
])

plt.imshow(reverse_preprocess(y));
```
